```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
%matplotlib inline
```

# Проектная работа по изучению счастья в мире

В ходе проектной работы мной будет исследован датасет https://www.kaggle.com/unsdsn/world-happiness в котором приведён ряд индексов, характеризующих социальные и экономические показатели жителей конкретных стран, из которых выводится общий уровень счастья. Данные взяты за 2016 год. Рассмотрим как распределено счастья по планете, а так же какой вклад вносят каждый из индексов в общий уровень счастья для различных регионов.

```python
df = pd.read_csv('/home/alex/Downloads/world-happiness-report/2016.csv')
df.head()
```

```python
df.info()
```

В первую очередь удалим Lower Confidence Interval и Upper Confidence Interval колонки, поскольку они лишь определяют доверительный интервал для уровня счастья и не несут в себе полезной информации.

```python
df = df.drop(['Lower Confidence Interval', 'Upper Confidence Interval'], axis=1)
df.info()
```

Теперь визуализируем уровень счастья стран на карте

```python
data = dict(type = 'choropleth', 
           locations = df['Country'],
           locationmode = 'country names',
           z = df['Happiness Score'], 
           text = df['Country'],
           colorbar = {'title':'Happiness'})
layout = dict(title = 'Global Happiness', 
             geo = dict(showframe = False, 
                       projection = {'type': 'mercator'}))
choromap3 = go.Figure(data = [data], layout=layout)
iplot(choromap3)
```

Теперь построим матрицу корреляции для определения наиболее значимых для уровня счастья показателей

```python
sns.heatmap(df.drop('Happiness Rank', axis=1).corr(), annot = True, cmap='inferno')
plt.show()
```

Как видим для всего датасета в целом с уровнем счастья сильнее всего коррелируют такие показатели как Economy, Health, Family. Теперь построим такие корреляционные матрицы для каждого региона, чтобы выяснить как изменяются веса каждого из показателей.

```python
df['Region'].unique()
```

```python
sns.heatmap(df.loc[df['Region'] == 'Western Europe'].drop('Happiness Rank', axis=1).corr(), annot = True, cmap='inferno')
plt.show()
```

Как видим, если рассматривать одну только восточную европу, то здесь совершенно другие показатели вносят наибольший вклад в уровень счастья людей. Это Trust, Freedom, Family

```python
sns.heatmap(df.loc[df['Region'] == 'Middle East and Northern Africa'].drop('Happiness Rank', axis=1).corr(), annot = True, cmap='inferno')
plt.show()
```

Для среднего востока и северной африки наиболее ценными являются Family, Economy, Freedom

```python
sns.heatmap(df.loc[df['Region'] == 'Latin America and Caribbean'].drop('Happiness Rank', axis=1).corr(), annot = True, cmap='inferno')
plt.show()
```

Примерно похожая ситуация в латинской америке. Это Economy, Health, Family

```python
sns.heatmap(df.loc[df['Region'] == 'Southeastern Asia'].drop('Happiness Rank', axis=1).corr(), annot = True, cmap='inferno')
plt.show()
```

Интересная картина в юго-восточной азии. Большой вес имеет Economy, Family, Health, что, примерно, как и везде. Однако Trust и Freedom здесь являются наименее значимыми, в отличии от европы.

```python
sns.heatmap(df.loc[df['Region'] == 'Central and Eastern Europe'].drop('Happiness Rank', axis=1).corr(), annot = True, cmap='inferno')
plt.show()
```

```python
sns.heatmap(df.loc[df['Region'] == 'Eastern Asia'].drop('Happiness Rank', axis=1).corr(), annot = True, cmap='inferno')
plt.show()
```
